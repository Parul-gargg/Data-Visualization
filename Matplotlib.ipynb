```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Matplotlib** **Basics**

```python
x = np.linspace(0,5,11)
x
y = x**3
y
```

```python
plt.plot(x,y)
plt.title('Hello')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
```


**Subplots** **in** **Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

```python
plt.subplot(6,6,1)
plt.subplot(6,6,2)
plt.subplot(6,6,3)
plt.subplot(6,6,7)
plt.subplot(6,6,9)
plt.subplot(6,6,5)
```

**Object** **Oriented** **way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(8,6), dpi = 100)
axis1 = fig.add_axes([0.4,0.4,0.6,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.6,.3,.3])
axis2.plot(x,y)
# here figure size values comes in between 0 and 1.
# 0.4 - left se kitna move krna hai
# 0.4 - bottom se kitna move krna hai
# 0.6 - size 60% is the width
# 0.8 - sizr 80% is the height
# axis 2 wala plot axis 1 wale plot par set kiya hai
```
